```python
from dputils.scrape import Scraper, Tag
```

```python
url = "https://en.wikipedia.org/wiki/Monkey_King"
sc = Scraper(url)
```

```python
sc.soup
```

```python
type(sc)
```

```python
type(sc.soup)
```

```python
print(sc.soup.text)
```

```python
pargraphs =sc.soup.findAll('p')
pargraphs
```

```python
data = ""
for item in pargraphs:
    text = item.text
    data+= text.strip() +"\n"
```

```python
print(data)
```

```python
# saving the data
with open("monkey_king.txt","w") as file:
    file.write(data.strip())
```
