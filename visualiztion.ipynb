```python
data = [1,2,3,4,5,6,8,10,12]
fig,ax = plt.subplots(figsize=(5,3))
ax.set_xlabel("no of elements",fontdict ={"fontsize": 10,"color":"red"})
ax.set_ylabel("y - value")
ax.grid(True)
ax.set_title("basics of visualization", loc="right")
plt.plot(data)
```
